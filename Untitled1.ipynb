```python
x = ev_car.iloc[:, [4, 14]].values
```

```python
#finding optimal number of clusters using the elbow method  
from sklearn.cluster import KMeans  
wcss_list= []  #Initializing the list for the values of WCSS  
  
#Using for loop for iterations from 1 to 10.  
for i in range(1, 11):  
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 42)  
    kmeans.fit(x)  
    wcss_list.append(kmeans.inertia_)  
plt.plot(range(1, 11), wcss_list)  
plt.title('The Elobw Method Graph')  
plt.xlabel('Number of clusters(k)')  
plt.ylabel('wcss_list')  
plt.show()  
```

```python
from sklearn.preprocessing import LabelEncoder

features = ['AccelSec', 'TopSpeed_KmH']

# Fit KMeans model
num_clusters = 4  # You can choose the number of clusters you want
kmeans = KMeans(n_clusters=num_clusters)
ev_car['Cluster'] = kmeans.fit_predict(ev_car[features])

# Set the style of the visualization
sb.set(style="whitegrid")

# Create a scatter plot based on 'Segment' and cluster centers
plt.figure(figsize=(12, 8))
sb.scatterplot(data=ev_car, x='AccelSec', y='TopSpeed_KmH', hue='Cluster', palette='viridis', s=100)
sb.scatterplot(x=kmeans.cluster_centers_[:, 0], y=kmeans.cluster_centers_[:, 1], color='red', marker='X', s=150)
plt.xlabel('Acceleration (0-100 km/h)')
plt.ylabel('Top Speed (km/h)')
plt.title('KMeans Clusters Based on Acceleration and Top Speed')
plt.legend(title='Cluster')
plt.tight_layout()

# Show the plot
plt.show()
```
